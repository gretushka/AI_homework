1. Посмотреть документацию к API GitHub, разобраться как вывести список репозиториев для конкретного пользователя, сохранить JSON-вывод в файле *.json.

```python
pip install requests
```

```python
import requests
import json

headers = {'User-Agent': 'Chrome/76.0.3809.100'}
name = 'gretushka'
url = f'https://api.github.com/users/{name}/repos'

response = requests.get(url, headers=headers).json()

with open('data.json', 'w', encoding='UTF-8') as f:
    json.dump(response, f)
    
for num, repo in enumerate(response):
    print(f"{num+1}. {repo.get('name')}")
```

2. Изучить список открытых API (https://www.programmableweb.com/category/all/apis). Найти среди них любое, требующее авторизацию (любого типа). Выполнить запросы к нему, пройдя авторизацию. Ответ сервера записать в файл.

```python
access_token=''
url = f'https://api.vk.com/method/groups.get?v=5.81&access_token={access_token}'
response = requests.get(url).json()
print(response)
```

```python
ids = response['response']['items']
ids
```

```python
url = f'https://api.vk.com/method/groups.getById?v=5.81&group_ids={ids}&access_token={access_token}'
response = requests.get(url).json()
print(response)
```
